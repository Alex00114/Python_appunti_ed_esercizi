```python
import pandas as pd
import xlrd 
```

```python
categorie = pd.read_excel('https://github.com/wtitze/3E/blob/main/BikeStores.xls?raw=true', "categories")
brand = pd.read_excel('https://github.com/wtitze/3E/blob/main/BikeStores.xls?raw=true',"brands")
prodotti =  pd.read_excel('https://github.com/wtitze/3E/blob/main/BikeStores.xls?raw=true',"products")
```

1. il catalogo complessivo dei prodotti. Visualizzare il nome del prodotto, il nome della marca e la categoria

```python
prodotti_brand = pd.merge(prodotti, brand, how = "outer", on = "brand_id")
prodotti_brand_categorie = pd.merge(prodotti_brand, categorie, how = "outer", on = "category_id")[["product_name","brand_name","category_name"]]
prodotti_brand_categorie
```

2. il numero di prodotti per ogni categoria. Visualizzare i nomi delle categorie e il numero dei prodotti.

```python
prod_categorie = pd.merge(prodotti, categorie, how = "inner", on = "category_id").groupby("category_name")[["product_name"]].count().reset_index()
prod_categorie
```

3. il numero di prodotti per ogni marca (brand). Visualizzare i nomi dei brand e il numero di prodotti

```python
prodotti_brand.groupby("brand_name", as_index = False)["product_name"].count()
```

4. i nomi dei prodotti della marca Trek

```python
trek = prodotti_brand[prodotti_brand["brand_name"] == "Trek"][["product_name"]]
trek
```

5. i prodotti delle categorie che hanno la parola 'Bikes' nel nome

```python
prod_categorie2 = pd.merge(prodotti, categorie, how = "inner", on = "category_id")
Bikes_prodotti = prod_categorie2[prod_categorie2["category_name"].str.contains("Bikes")][["product_name"]]
Bikes_prodotti
```

6. la categoria che ha il prodotto che costa di più

```python
prezzo_alto = pd.merge(prodotti, categorie, how = "inner", on = "category_id").groupby("category_name",as_index=False)["list_price"].max().sort_values(by="list_price").tail(1)
prezzo_alto 
```

7. i brand con almeno 100 prodotti. Viaulizzare il nome del brand e il numero di prodotti

```python
prodotti_ordinati = pd.merge(prodotti, brand, how = "inner", on = "brand_id").groupby("brand_name", as_index = False)["product_name"].count().sort_values(by="product_name",ascending = False)
cento_prodotti = prodotti_ordinati.loc[prodotti_ordinati["product_name"] >= 100]
cento_prodotti
```
