```python
import pandas as pd
import matplotlib.pyplot as plt
import squarify as sqf
```

a. Crei un dataframe contenente le informazioni riportate nella tabella sottostante, relativa al numero di
ordini (order_numbers) seguiti da vari commessi (last_name) di un negozio.

```python
d = {"last_name" : ["Boyer", "Copeland", "Daniel", "Serrano", "Terrell", "Vargas"],
      "order_numbers" : [185, 59, 196, 58, 24, 28]
      }

df = pd.DataFrame(d)
df
```

b. Visualizzi un grafico a barre verticali per rappresentare i dati.

```python
fig = plt.figure(figsize = (8,6))
ax = plt.axes()

ax.bar(df["last_name"], df["order_numbers"])
plt.show()
```

c. Visualizzi lo stesso grafico ma con altri colori delle barre, un titolo appropriato al grafico e i nomi degli
assi.

```python
fig = plt.figure(figsize = (8,7))
ax = plt.axes()
ax.set_xlabel('last_name', fontsize = 20)
ax.set_ylabel('order_numbers', fontsize = 20)
fig.suptitle('ISTOGRAMMA', fontsize = 30)

ax.bar(df["last_name"], df["order_numbers"], color = "c")
plt.show()
```

d. Visualizzare lo stesso grafico ma con le barre in ordine decrescente.

```python
fig = plt.figure(figsize = (8,7))
ax = plt.axes()
ax.set_xlabel('last_name', fontsize = 20)
ax.set_ylabel('order_nubers', fontsize = 20)
fig.suptitle('ISTOGRAMMA', fontsize = 30)

ax.bar(df.sort_values("order_numbers", ascending = False)["last_name"], df["order_numbers"].sort_values(ascending = False), color = "c")
plt.show()
```

e. Visualizzi un grafico a barre orizzontali per rappresentare i dati.

```python
fig = plt.figure(figsize = (8,7))
ax = plt.axes()
ax.set_ylabel('last_name', fontsize = 20)
ax.set_xlabel('order_numbers', fontsize = 20)
fig.suptitle('ISTOGRAMMA', fontsize = 30)

ax.barh(df["last_name"], df["order_numbers"], color = "c")
plt.show()
```

f. Crei un nuovo dataframe contenente la percentuale di ordini consegnati seguiti da ogni componente
dello staff. Visualizzare il cognome del componente dello staff e la percentuale di ordini. Ordinare in
ordine alfabetico sul cognome del componente dello staff.

```python
d = {"last_name" : ["Copeland", "Serrano", "Boyer", "Daniel", "Vargas", "Terrell"],
      "order_perc" : [9.242144, 10.720887, 34.195933, 36.229205, 5.175601, 4.436229]
      }

df = pd.DataFrame(d)
df
```

g. Visualizzi un grafico a torta per rappresentare i dati estratti.

```python
perc = list(df["order_perc"])
nomi = list(df["last_name"])

fig = plt.figure(figsize = (10,10))
ax = plt.axes()
plt.rcParams.update({'font.size': 15})  
ax.pie(perc, labels=nomi, autopct='%1.2f%%')
plt.show
```

h. Visualizzi lo stesso grafico ma con altri colori delle percentuali e un titolo appropriato al grafico

```python
perc = list(df["order_perc"])
nomi = list(df["last_name"])
cols = ["m","c","r","g","b", "y"]

fig = plt.figure(figsize = (10,10))
ax = plt.axes()
plt.rcParams.update({'font.size': 15})  
fig.suptitle('Grafico a Torta', fontsize = 30)

ax.pie(perc, labels=nomi, colors = cols, autopct='%1.2f%%')
plt.show()
```

i. Rappresentare le stesse informazioni utilizzando come tipo di grafico il TreeMap (google it)

```python
perc = list(df["order_perc"])
nomi = list(df["last_name"])
sqf.plot(sizes = perc, label = nomi, alpha = 0.6)
plt.axis("off")
plt.show()
```
