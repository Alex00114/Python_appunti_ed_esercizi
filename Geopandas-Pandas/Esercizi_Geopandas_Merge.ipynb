```python
province = geopandas.read_file("/content/drive/MyDrive/Colab Notebooks/Limiti01012021_g/Limiti01012021_g/ProvCM01012021_g")
regioni = geopandas.read_file("/content/drive/MyDrive/Colab Notebooks/Limiti01012021_g/Limiti01012021_g/Reg01012021_g")
comuni = geopandas.read_file("/content/drive/MyDrive/Colab Notebooks/Limiti01012021_g/Limiti01012021_g/Com01012021_g")
```

- Visualizzare il nome della provincia e il nome della regione

```python
prov_regioni = pd.merge(province, regioni, how = "inner", on = "COD_REG")
prov_regioni[["DEN_PROV", "DEN_REG"]]
```

- Visualizzare il nome della provincia e il nome del comune

```python
prov_comuni = pd.merge(province, comuni, how = "inner", on = "COD_PROV")
prov_comuni[["DEN_PROV", "COMUNE"]]
```

- Visualizzare il nome del comune e il nome della regione

```python
com_regioni = pd.merge(regioni, comuni, how = "inner", on = "COD_REG")
com_regioni[["COMUNE", "DEN_REG"]]
```

- Visualizzare numero di comuni per ogni provincia


```python
prov_comuni_groupby = prov_comuni.groupby("DEN_PROV")["COMUNE"].count()
prov_comuni_groupby
```

- Visualizzare numero di comuni per ogni regione

```python
com_regioni_groupby = com_regioni.groupby("DEN_REG")["COMUNE"].count()
com_regioni_groupby
```

- Visualizzare numero di province per ogni regione


```python
prov_regioni_groupby = prov_regioni.groupby("DEN_REG")["DEN_PROV"].count()
prov_regioni_groupby
```
